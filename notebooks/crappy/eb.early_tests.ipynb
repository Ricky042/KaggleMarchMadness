# Early Testing with Polars

```python
import polars as pl
from pathlib import Path

DATA_DIR = Path("../data/landing/march-machine-learning-mania-2025")

mteams = pl.read_csv(DATA_DIR / "MTeams.csv")
wteams = pl.read_csv(DATA_DIR / "WTeams.csv")

mseasons = pl.read_csv(DATA_DIR / "MSeasons.csv")
wseasons = pl.read_csv(DATA_DIR / "WSeasons.csv")

mresults = pl.read_csv(DATA_DIR / "MRegularSeasonDetailedResults.csv")
wresults = pl.read_csv(DATA_DIR / "WRegularSeasonDetailedResults.csv")

mseeds = pl.read_csv(DATA_DIR / "MNCAATourneySeeds.csv")
wseeds = pl.read_csv(DATA_DIR / "WNCAATourneySeeds.csv")

all_results = pl.concat([mresults, wresults])

def replace_team_ids_with_names(results, teams):
    results = results.join(
        teams.select(["TeamID", "TeamName"]).rename({"TeamID":"WTeamID", "TeamName":"WTeamName"}),
        on="WTeamID", how="left"
    )
    
    results = results.join(
        teams.select(["TeamID", "TeamName"]).rename({"TeamID":"LTeamID", "TeamName":"LTeamName"}),
        on="LTeamID", how="left"
    )
    return results

mresults_named = replace_team_ids_with_names(mresults, mteams)
wresults_named = replace_team_ids_with_names(wresults, wteams)

name_change = pl.concat([mresults_named, wresults_named])
display(name_change)

def clean_seed(df):
    return df.with_columns([
        pl.col("Seed").str.slice(1, 2).cast(pl.Int32).alias("SeedNum")
    ])

mseeds = clean_seed(mseeds)
wseeds = clean_seed(wseeds)

```
